## 네이버 금융 코스피

```python
page_no=1
page_url=f"https://finance.naver.com/sise/sise_index_day.naver?code=KOSPI&page={page_no}"
page_url
```

```python
import requests

source=requests.get(page_url).text
source
```

```python
import bs4
source=bs4.BeautifulSoup(source)
source
```

```python
print(source.prettify())
```

```python
#source에서 'td'태그에 클래스가 number_1인 것들만 뽑아오기
source.find_all('td',class_='number_1')
```

```python
dates=source.find_all("td",class_="date")
date_list=[]
for date in dates:
    date_list.append(date.text)
date_list
```

```python
prices=source.find_all("td",class_="number_1")
price_list=[]
for price in prices[::4]:
    price_list.append(price.text)

price_list
```

```python
len(date_list)
```

```python
len(price_list)
```

```python
#맨끝버튼
last_url=source.find_all("td",class_="pgRR")[0].find_all("a")[0]["href"]
last_url
```

```python
#마지막 페이지를 알기 위해 마지막 글자 떼어내기
last_page=int(last_url.split('&page=')[-1])
last_page
```

#### 합쳐서 구현

```python
date_list=[]
price_list=[]

for page_no in range(1,last_page+1):
    page_url=f"https://finance.naver.com/sise/sise_index_day.naver?code=KOSPI&page={page_no}"

    source=requests.get(page_url).text
    source=bs4.BeautifulSoup(source)

    dates=source.find_all("td",class_="date")

    for date in dates:
        date_list.append(date.text)

    prices=source.find_all("td",class_="number_1")

    for price in prices[::4]:
        price_list.append(price.text)
```

```python
1429*6
```

```python
len(date_list),len(price_list)
```

```python
import pandas as pd

df=pd.DataFrame({"date":date_list,"price":price_list})
df
```

```python
df.to_excel("kpi.xlsx",index=False)
```
